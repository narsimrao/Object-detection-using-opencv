```python
from imageai.Detection import ObjectDetection
import os
import cv2 as cv2
from google.colab.patches import cv2_imshow

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath('/content/drive/MyDrive/Colab Notebooks/yolo.h5')
detector.loadModel()
detections = detector.detectCustomObjectsFromImage(input_image="/content/drive/MyDrive/Colab Notebooks/020160322000524.jpg", 
                                             output_image_path="/content/imagenew.jpg", 
                                             minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
img = cv2.imread('/content/imagenew.jpg')
cv2_imshow(img)
```
